### **Q1** Sampling theorem and aliasing

<div style='background:#F0F0F0'>**1** Create a band-limited white noise timeseries with $N=1000$ entries, $dt=0.1 s$, but with all frequencies greater than 0.5 Hz filtered.  Plot the power spectrum to make sure you got it correctly...
</div>

```python
N = 1000
x = np.random.rand(N)
t = np.arange(0, 0.1*N, 0.1)

#lowpass filter
def filter(x):
    dt = 0.1
    B = 0.5
    fp=B*0.7
    fs = B*0.9
    
    fNyq=1/2./dt
    n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 40.)
    b,a=signal.ellip(n,.1,40.,fn)
    xLF = signal.filtfilt(b,a,x)
    return xLF, dt, B

xLF, dt, B = filter(x)
gx, fx = mlab.psd(x, NFFT = len(x), Fs = 1./dt)
gfxLF, ffxLF = mlab.psd(xLF, NFFT = len(x), Fs = 1./dt)

fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(2, 1, 1)
ax.plot(t, x, color ='grey', label='Raw (White) Data')
ax.plot(t, xLF, color = 'black', label = 'Filtered Data')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Raw White Noise')
ax.set_title('Raw Noise Data')
ax.legend()

ax1 = fig.add_subplot(2, 1, 2)
ax1.loglog(fx, gx, color = 'grey', label= "Spectrum of x")
ax1.loglog(ffxLF, gfxLF, color = 'black', label= "Filtered Spectrum of x")
ax1.axvline(0.5, color='r', linestyle='dashed', linewidth=2)
ax1.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Power Spectrum $G_{xx}$ [$V^{2}$Hz]')
ax1.set_title('Power Spectrum')
ax1.legend()

plt.tight_layout()
```

<div style='background:#F0F0F0'>**2** Now decimate by a factor of 4 to get a time series sampled at 2.5 Hz.  Show that you can recover the 10 Hz time series using the Sampling Theorem. </div>

```python
fs = 2.5
dts = 1./fs
tn = t[::4]
xn = xLF[::4]
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(t, xLF, linewidth = 3.0, color = 'black', label = 'Filtered Data')

n = np.arange(N/4)
xreco=1.*xLF[:N/4]
for i in range(N/4):
    xreco[i] = np.sum(xn[:N]*np.sinc(((1./dts)*tn[i])-n)) 
    

ax.plot(tn,xreco, linewidth = 1.0, label='Reconstructed Filtered Data at 2.5 Hz')
ax.set_xlabel('Time [s]')
ax.set_ylabel('$x_{recon}$')
ax.set_title('Reconstruction of Filtered Data')
ax.legend()
```

When the filter data was reconstructed with a decimation value of 4 to get a time series sampled at 2.5Hz the resultant recontructed filtered data follows the original filtered time series.  We can say that frequencies are preserved and the 10 Hz time series is recovered.

<div style='background:#F0F0F0'>**3** Now decimate by a factor of 40 to get a time series sampled at 0.25 Hz.  Show that you cannot recover the 10 Hz time series using the Sampling Theorem. </div>

```python
fs = 0.25
dts2 = 1./fs
tn2 = t[::40]
xn2 = xLF[::40]
fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(t, xLF, linewidth = 3.5, color = 'black', label = 'Filtered Data')

n = np.arange(N/40)
xreco2=1.*xLF[:N/40]
for i in range(N/40):
    xreco2[i] = np.sum(xn2[:N]*np.sinc(((1./dts2)*tn2[i])-n)) 
    

ax.plot(tn2,xreco2, linewidth = 1.0, label='Reconstructed Filtered Data at 0.25 Hz')
ax.set_xlabel('Time [s]')
ax.set_ylabel('$x_{recon}$')
ax.set_title('Reconstruction of Filtered Data')
ax.legend()
```

When a decimation value of 40 is used on the data set the resultant filtered data does not follow the path of the inital filtered data.  The decimation to a frequency of 0.25Hz does not preserve the filtered data and thus we cannot recover the 10Hz time series.

## Q2: Cubic Spline

Using the data below...

```python
t2 = np.array([0.,2.5,2.7,3.4,4.2,6.5,9.0])
x2= np.cos(t2*2*np.pi/3)
```

<div style='background:#F0F0F0'>**1** Perform a cubic spline interpolation onto an even grid from 0 to 9.0 by 
  1. creating the matrix from the notes
  2. inverting and solving for the cubic co-efficients
  3. using those polynomials to evaluate the interpolated data.
Compare with using `scipy.interpolate.interp1d`.
Comment on any similarities or differences between the two results.
</div>

```python
N2 = len(x2)-1
n = np.arange(0, N2, 1)


delta0 = t2[1]-t2[0]
delta1 = t2[2]-t2[1]
delta2 = t2[3]-t2[2]
delta3 = t2[4]-t2[3]
delta4 = t2[5]-t2[4]
delta5 = t2[6]-t2[5]
delta = np.array([[delta0],[delta1], [delta2], [delta3], [delta4], [delta5]])

#1 create matrix
matrix = np.array([[1, 0, 0, 0, 0, 0, 0],
          [delta0,(2*(delta0+delta1)), delta1, 0, 0, 0, 0],
          [0,delta1,(2*(delta1+delta2)), delta2, 0, 0, 0],
          [0, 0, delta2,(2*(delta2+delta3)), delta3, 0, 0],
          [0, 0, 0, delta3,(2*(delta3+delta4)), delta4, 0],
          [0, 0, 0, 0, delta4,(2*(delta4+delta5)), delta5],
          [0, 0, 0, 0, 0, 0, 1]])

dx0 = x2[1] - x2[0]
dx1 = x2[2] - x2[1] 
dx2 = x2[3] - x2[2]
dx3 = x2[4] - x2[3]
dx4 = x2[5] - x2[4]
dx5 = x2[6] - x2[5]
dx = np.array([[dx0], [dx1], [dx2], [dx3], [dx4], [dx5]])

diffmatrix = np.array([[0],
                       [(dx1/delta1) - (dx0/delta0)], 
                       [(dx2/delta2) - (dx1/delta1)], 
                       [(dx3/delta3) - (dx2/delta2)], 
                       [(dx4/delta4) - (dx3/delta3)], 
                       [(dx5/delta5) - (dx4/delta4)], 
                       [0]])
diffmatrix = 3*diffmatrix

#invert and solve for cubic coefficients
c = np.linalg.solve(matrix, diffmatrix) 
d = np.zeros(N2)
b = np.zeros(N2)
for i in range (0, N2-2, 1): 
    d[i] = (2*(c[i+1] - c[i]))/(6*delta[i]) 
d[N2-1]= (-2*c[N2-1])/(6*delta[N2-1])
for i in range (0, N2-1, 1): 
    b[i] = (x2[i+1] - x2[i] - (delta[i]**2*c[i])-((delta[i]**3)*d[i]))/delta[i]   

#use polynomial to evaluate interpolated data
tint = 10
t01 = np.linspace(0, 2.5, tint)
t12 = np.linspace(2.5, 2.7, tint)
t23 = np.linspace(2.7, 3.4, tint)
t34 = np.linspace(3.4, 4.2, tint)
t45 = np.linspace(4.2, 6.5, tint)
t56 = np.linspace(6.5, 9.0, tint)
tarray = np.array([t01, t12, t23, t34, t45, t56])
q = np.zeros([N2-1, tint])
for i in range (0, N2-1, 1): 
    tint= tarray[i]
    for j in range (0, 10, 1):
        q[i,j] = x2[i] + b[i]*(tint[j]-t2[i])+c[i]*(tint[j]-t2[i])**2+d[i]*(tint[j]-t2[i])**3
q = q.flatten()
tarray = tarray.flatten()

#scipy.interpolate.interp1d 
data=interp1d(t2,x2,kind='cubic')
ts = np.linspace(0, 9.0, 100)
xs = data(ts)

#plots and comparisons
fig = plt.figure(figsize = (15, 5))
ax2 = fig.add_subplot(1, 1, 1)
ax2.plot(tarray[:-10], q, label= "Worked Interpolation")
ax2.plot(ts, xs, label = "Scipy Function for interp1d")
ax2.plot(t2, x2, label = "Raw Data")
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Position')
ax2.set_title('Cubic Spline Interpolation of Data ')
ax2.legend(loc = 3)
```
