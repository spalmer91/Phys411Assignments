### **Q1** Frequency-domain filtering

<div style='background:#F0F0F0'>  Consider a time series with a sharp jump in the middle as in the notes:
</div>

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab

N=1001
dt=1.
t=np.arange(0,N*dt,dt)

x = np.cumsum(np.random.randn(N)) # red noise
f0=0.03 # Hz
x = x+ 40.*np.sin(f0*2*np.pi*t) # + sine wave
x=x-np.mean(x)
noise = 60.*np.random.randn(N)
xn = x+noise
xn[N/2:]=xn[N/2:]-2000.
# remove the mean!!
xn = xn-np.mean(xn)
```

<div style='background:#F0F0F0'>**1** Implement a low-pass filter of the data in the frequency domain, with a cutoff frequency of $f_c=0.06\ Hz$.</div>
   
  1. plot the signal and the smoothed signal (suggest zooming in on the step).
  2. plot the spectral response $|H(f)|$
  3. plot the phase response $\mathrm{angle}(H(f))$.

<div style='background:#F0F0F0'>
Be sure to comment on how good your filter is.
</div>

```python
#part 1
X = np.fft.fft(xn)
#shape of xn
N = np.shape(xn)[0]
#fourier frequency space
ff = np.linspace(0,1./dt,N)
#mask freqeuncies not in region of study
X[(ff>0.06) & (ff<(1./dt-0.06))]=1.e-15
# There is a strange bug in psd that makes adding this little bit of noise necessary.
y = np.real(np.fft.ifft(X))+1.e-17*np.random.randn(N)

#part 2 and 3
#psd and csd calculations
nfft = 256
pxn, fxn = mlab.psd(xn, Fs = 1./dt, NFFT = nfft)
py, fy = mlab.psd(y, Fs = 1./dt, NFFT = nfft)
pxny, fxny = mlab.csd(xn, y, Fs = 1./dt, NFFT = nfft)
hfabs= np.abs(np.sqrt(py/pxn))
angle = np.angle(pxny)

#plots
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(4, 1, 1)
ax.plot(t, xn, label = 'Signal')
ax.plot(t, y, label = 'Filtered Signal and Noise')
ax.set_xlabel ('Time [s]')
ax.set_ylabel('Signal')
ax.set_title('Signal Filtration')
ax.legend()

ax = fig.add_subplot(4, 1, 2)
ax.plot(t[475:525], xn[475:525], label = 'Signal and Noise in Time 475 to 525s')
ax.plot(t[475:525], y[475:525], label = 'Filtered Signal and Noise in Time 475 to 525s')
ax.set_xlabel ('Time [s]')
ax.set_ylabel('Signal')
ax.set_title('Signal Filtration')
ax.legend()

ax1 = fig.add_subplot(4, 1, 3)
ax1.loglog(fy, hfabs, label ='Spectral Response')
ax1.set_xlabel ('Frequency [Hz]')
ax1.set_ylabel('$|H(f)|$')
ax1.set_title('Spectral Response')

ax2 = fig.add_subplot(4, 1, 4)
ax2.semilogx(fy, angle, label ='Phase Response')
ax2.set_xlabel ('Frequency [Hz]')
ax2.set_ylabel('Phase')
ax2.set_title('Phase Response')

plt.tight_layout()
```

This simple filter is not exceptional in procuding a smoothed singal with noise. We can see that when smoothed the data using a spectral response is well correlated at low frequencies and disippates off at higher frequencies.  As well the phase is initially in sequence and then become uncorrelated in the higher frquencies.  The breaking point is a frequency of about 0.0055Hz.  Thus we can say that higher frequencies are being filtered well.

<div style='background:#F0F0F0'>**2** Now repeat the above but apply a Hanning window to the frequency window *before* you apply it to the FFT of $x_n$.  Comment on how well the filter is doing when you do this.  (Its helpful to keep the previous result up)
</div>

```python
#hanning window
window = np.hanning(int(len(ff)*0.06*2))

Xw = np.fft.fft(xn)
#shape of xn
N = np.shape(xn)[0]
#mask freqeuncies not in region of study
Xw[(ff>0.06) & (ff<(1./dt-0.06))]=1.e-15
#hanning window the beginning
Xw[:0.06*len(ff)] = Xw[:0.06*len(ff)]*window[0.06*len(ff):]
#hanning window the end
Xw[-0.06*len(ff):] = Xw[-0.06*len(ff):]*window[:0.06*len(ff)]
# There is a strange bug in psd that makes adding this little bit of noise necessary.
yw = np.real(np.fft.ifft(Xw))+1.e-17*np.random.randn(N)

#part 2 and 3
#psd and csd calculations
nfft = 256
pxn, fxn = mlab.psd(xn, Fs = 1./dt, NFFT = nfft)
pyw, fyw = mlab.psd(yw, Fs = 1./dt, NFFT = nfft)
pxnyw, fxny = mlab.csd(xn, yw, Fs = 1./dt, NFFT = nfft)
hfabsw= np.abs(np.sqrt(pyw/pxn))
anglew = np.angle(pxnyw)

#plots
fig = plt.figure(figsize = (15, 10))
ax = fig.add_subplot(4, 1, 1)
ax.plot(t, xn, label = 'Signal')
ax.plot(t, y, label = 'Filtered Signal and Noise')
ax.plot(t, yw, label = 'Filtered Signal and Noise with Hanning Window')
ax.set_xlabel ('Time [s]')
ax.set_ylabel('Signal')
ax.set_title('Signal Filtration')
ax.legend()

ax = fig.add_subplot(4, 1, 2)
ax.plot(t[475:525], xn[475:525], label = 'Signal and Noise')
ax.plot(t[475:525], y[475:525], label = 'Filtered Signal and Noise')
ax.plot(t[475:525], yw[475:525], label = 'Filtered Signal and Noise with Hanning Window')
ax.set_xlabel ('Time [s]')
ax.set_ylabel('Signal')
ax.set_title('Signal Filtration')
ax.legend()

ax1 = fig.add_subplot(4, 1, 3)
ax1.loglog(fy, hfabs, label ='Spectral Response')
ax1.loglog(fyw, hfabsw, label ='Spectral Response with Hanning Window')
ax1.set_xlabel ('Frequency [Hz]')
ax1.set_ylabel('$|H(f)|$')
ax1.set_title('Spectral Response')
ax1.legend()

ax2 = fig.add_subplot(4, 1, 4)
ax2.semilogx(fy, angle, label ='Phase Response')
ax2.semilogx(fyw, anglew, label ='Phase Response with Hanning Window')
ax2.set_xlabel ('Frequency [Hz]')
ax2.set_ylabel('Phase')
ax2.set_title('Phase Response')
ax2.legend()

plt.tight_layout()
```

When the hanning window is applied to the source we can see that the spectral respose drops off at sooner high frequencies and the phase response is less correlated sooner (at approximately 0.0014Hz).  This means that the filtering is acting on a smaller group of high frequencies and the resultant product is a smoother jump at approximately 500 seconds and smoother data behind and infront of the 500s as can be seen in the second plot above of the signal filtration zoomed in.

<div style='background:#F0F0F0'>**3** So the Hanning window falls off a *lot* earlier and more slowly than a step function, so adjust the cutoff frequency to get a reasonable filter that preserves the amplitude of the sine wave at $f=0.03\ \mathrm{Hz}$.  Comment on the pros and cons of this approach.  
</div>

```python
def cutofffreq(fc):
    #f must be much much larger! 
    # attempt 0.09, 0.1, 0.11, 0.12, 0.13, 0.14
    #hanning window
    window = np.hanning(int(len(ff)*fc*2))

    Xw = np.fft.fft(xn)
    #shape of xn
    N = np.shape(xn)[0]
    #mask freqeuncies not in region of study
    Xw[(ff>fc) & (ff<(1./dt-fc))]=1.e-15
    #hanning window the beginning
    Xw[:fc*len(ff)] = Xw[:fc*len(ff)]*window[fc*len(ff):]
    #hanning window the end
    Xw[-fc*len(ff):] = Xw[-fc*len(ff):]*window[:fc*len(ff)]
    # There is a strange bug in psd that makes adding this little bit of noise necessary.
    yw = np.real(np.fft.ifft(Xw))+1.e-17*np.random.randn(N)

    #part 2 and 3
    #psd and csd calculations
    nfft = 256
    pxn, fxn = mlab.psd(xn, Fs = 1./dt, NFFT = nfft)
    pyw, fyw = mlab.psd(yw, Fs = 1./dt, NFFT = nfft)
    pxnyw, fxny = mlab.csd(xn, yw, Fs = 1./dt, NFFT = nfft)
    hfabsw= np.abs(np.sqrt(pyw/pxn))
    anglew = np.angle(pxnyw)
    return yw, fyw, pxnyw, hfabsw, anglew
yw6, fyw6, pxnyw6, hfabsw6, anglew6 = cutofffreq(0.06)
yw9, fyw9, pxnyw9, hfabsw9, anglew9 = cutofffreq(0.09)
yw0, fyw0, pxnyw0, hfabsw0, anglew0 = cutofffreq(0.10)
yw1, fyw1, pxnyw1, hfabsw1, anglew1 = cutofffreq(0.11)
yw2, fyw2, pxnyw2, hfabsw2, anglew2 = cutofffreq(0.12)
yw3, fyw3, pxnyw3, hfabsw3, anglew3 = cutofffreq(0.13)
yw4, fyw4, pxnyw4, hfabsw4, anglew4 = cutofffreq(0.14)
#plots
fig = plt.figure(figsize = (15, 10))
ax = fig.add_subplot(4, 1, 1)
ax.plot(t, xn, label = 'Signal')
ax.plot(t, yw6, label = 'Cut off Frequency 0.06')
ax.plot(t, yw9, label = 'Cut off Frequency 0.09')
ax.plot(t, yw0, label = 'Cut off Frequency 0.10')
ax.plot(t, yw1, label = 'Cut off Frequency 0.11')
ax.plot(t, yw2, label = 'Cut off Frequency 0.12')
ax.plot(t, yw3, label = 'Cut off Frequency 0.13')
ax.plot(t, yw4, label = 'Cut off Frequency 0.14')
ax.set_xlabel ('Time [s]')
ax.set_ylabel('Signal Filtered with Cut off Frequency')
ax.set_title('Signal Filtration')
ax.legend()

ax = fig.add_subplot(4, 1, 2)
ax.plot(t[475:525], xn[475:525], label = 'Signal and Noise')
ax.plot(t[475:525], yw6[475:525], label = 'Filtered Signal and Noise with Hanning Window fc = 0.06')
#ax.plot(t[475:525], yw9[475:525], label = 'Filtered Signal and Noise with Hanning Window fc = 0.09')
#ax.plot(t[475:525], yw0[475:525], label = 'Filtered Signal and Noise with Hanning Window fc = 0.10')
#ax.plot(t[475:525], yw1[475:525], label = 'Filtered Signal and Noise with Hanning Window fc = 0.11')
ax.plot(t[475:525], yw2[475:525], label = 'Filtered Signal and Noise with Hanning Window fc = 0.12')
#ax.plot(t[475:525], yw3[475:525], label = 'Filtered Signal and Noise with Hanning Window fc = 0.13')
#ax.plot(t[475:525], yw4[475:525], label = 'Filtered Signal and Noise with Hanning Window fc = 0.14')
ax.set_xlabel ('Time [s]')
ax.set_ylabel('Signal')
ax.set_title('Signal Filtration')
ax.legend()

ax1 = fig.add_subplot(4, 1, 3)
ax1.loglog(fy, hfabs, label ='Spectral Response')
ax1.loglog(fyw6, hfabsw6, label = 'Cut off Frequency 0.06')
#ax1.loglog(fyw9, hfabsw9, label = 'Cut off Frequency 0.09')
#ax1.loglog(fyw0, hfabsw0, label = 'Cut off Frequency 0.10')
#ax1.loglog(fyw1, hfabsw1, label = 'Cut off Frequency 0.11')
ax1.loglog(fyw2, hfabsw2, label = 'Cut off Frequency 0.12')
#ax1.loglog(fyw3, hfabsw3, label = 'Cut off Frequency 0.13')
#ax1.loglog(fyw4, hfabsw4, label = 'Cut off Frequency 0.14')
ax1.set_xlabel ('Frequency [Hz]')
ax1.set_ylabel('$|H(f)|$')
ax1.set_title('Spectral Response with Hanning Window')
ax1.legend()

ax2 = fig.add_subplot(4, 1, 4)
ax2.semilogx(fy, anglew, label = 'Phase Response')
ax2.semilogx(fyw6, anglew6, label = 'Cut off Frequency 0.06')
#ax2.semilogx(fyw9, anglew9, label = 'Cut off Frequency 0.09')
#ax2.semilogx(fyw0, anglew0, label = 'Cut off Frequency 0.10')
#ax2.semilogx(fyw1, anglew1, label = 'Cut off Frequency 0.11')
ax2.semilogx(fyw2, anglew2, label = 'Cut off Frequency 0.12')
#ax2.semilogx(fyw3, anglew3, label = 'Cut off Frequency 0.13')
#ax2.semilogx(fyw4, anglew4, label = 'Cut off Frequency 0.14')
ax2.set_xlabel ('Frequency [Hz]')
ax2.set_ylabel('Phase')
ax2.set_title('Phase Response with Hanning Window')
ax2.legend()

plt.tight_layout()
```

## Q2: Implementing a bandpass filter

<div style='background:#F0F0F0'>**1** Create a red-noise process with 20000 data points, $dt=1\ s$.  Using a **truncated Dirichlet Kernel**, create a **bandpass** filter centered at $f_c=0.1\ Hz$ that has a passband that is approximately 0.1 Hz wide. Choose a filter length that is long enough to accomodate 0.1 Hz. 
</div>

   1. Plot the unshifted and shifted filters versus $k$, the time index.
   2. Plot the spectra, the response function $H^2(f)$ and the phase response
   3. Comment on your bandpass and if it indeed passed the right frequencies.

HINT: when you make the Dirichlet Kernel filter make sure that the $k=0$ case is treated "correctly" because $sin(0)=0$ and you will get a discontinuity.  Calculate what $h_0$ should be from [L'Hôpital's_rule](http://en.wikipedia.org/wiki/L'Hôpital's_rule).

```python
N2 = 200000
dt = 1.
t = np.arange(0, N2*dt, dt)
x2 = np.cumsum(np.random.rand(N2))
#remove mean
x2 = x2 - np.mean(x2)
# cut off frequency
fc = 0.1
M = fc*2*N2+1
mc = 0.1*N2
k = np.array(range(-N2/2+1,N2/2))
m = k*1.
h = (1./N2)*np.sin(np.pi*(k+0.000001)*M/N2)/np.sin(np.pi*(k+0.000001)/N2)

filtlen =  50
h = h[abs(k)<filtlen+1]
#k must be same length
k = k[abs(k)<filtlen+1]
hshift = 2*h*np.cos(2*np.pi*(k)*mc/N2)

#plot
fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(k, h, label='Unshifted Filters')
ax.plot(k, hshift, label='Shifted Filters')
ax.set_xlabel('Time Index, k')
ax.set_ylabel('$h_k$')
ax.set_title('Truncated Dirichlet Kernel with bandpass filter centered at $f_c$ = 0.1Hz')
ax.legend()
```

```python
nfft2 = N2/10 #from example N = 256*10 and your nfft was 256...
y2 = np.convolve(x2, h, mode = 'same')

pxx2, fx2 = mlab.psd(x2, Fs=1., NFFT = nfft2, noverlap = nfft2/2)
pyy2, fy2 = mlab.psd(y2, Fs=1., NFFT = nfft2, noverlap = nfft2/2)
# h squared value
H2 = pyy2/pxx2
# angle response
pxy2, fxy2 = mlab.csd(x2, y2, Fs=1., NFFT = nfft2)
angle2 = np.angle(pxy2)

fig = plt.figure(figsize = (15, 15))
#spectra
ax = fig.add_subplot(3, 1, 1)
ax.loglog(fx2, pxx2, label='Spectra of Raw')
ax.loglog(fy2, pyy2, label='Spectra of Filtered')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('$G_{xx}$ or $G_{yy}$ [$V^{2}s$]')
ax.set_title('Spectra')
ax.legend()
#response function
ax1 = fig.add_subplot(3, 1, 2)
ax1.loglog(fy2, H2, label='Response Function')
ax1.set_xlabel('f [Hz]')
ax1.set_ylabel('$|H|^{2}$')
ax1.set_title('Response Function')
ax1.legend()
#phase response
ax2 = fig.add_subplot(3, 1, 3)
ax2.semilogx(fxy2, angle2, label='Phase')
ax2.set_xlabel('Frequency [hZ]')
ax2.set_ylabel('Phase')
ax2.set_title('Phase Response')
ax2.legend()

plt.tight_layout()
```

As can be seen the filtering was effective in filtering frequenices higher than 0.1Hz and after that there is variations after the 0.1 Hz. It did indeed pass the correct frequencies.

<div style='background:#F0F0F0'> **2** Using your band passed signal, plot the *amplitude* of the daily cycle as a function of time, smoothed over a suitable time period. Compare with the raw timeseries to make sure your amplitudes make sense. Comment on the annual variation in the daily amplitude and if it makes sense with respect to the seasons.
</div>

```python
#raw data
tdc = np.arange(0, len(dc)/(60.*24.*365.), 1/(60.*24.*365.))
#filtered data
#daily
N = np.size(yl)/(60.*24.) #number of days
a = 0
b = 1
ydaily = np.zeros(N+1)
tdaily = np.arange(0,N,1)
for i in range (0, int(N+1), 1):
    low = a
    high = 60*24*b
    rangeyl = yl[low:high]
    maxyl = np.max(rangeyl)
    minyl = np.min(rangeyl)
    ampyl = maxyl-minyl
    ydaily[i] = ampyl
    a = a+60*24
    b= b+1

fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(2, 1, 1)
ax.plot(tdc, dc, label='Raw Deep Cove Data')
ax.plot(tdc, yl, label= 'Filtered Deep Cove Data')
ax.set_xlabel('Time (years)')
ax.set_ylabel('Temperature [$^\circ$ C]')
ax.set_title('Deep Cove Data with Mean Removed')
ax.legend()

ax1 = fig.add_subplot(2, 1, 2)
ax1.plot(tdaily, ydaily, label= 'Filtered Deep Cove Data')
ax1.set_xlabel('Time (days)')
ax1.set_ylabel('Temperature [$^\circ$ C]')
ax1.set_title('Daily Amplitude of Deep Cove Data with Mean Removed')
ax1.legend

plt.tight_layout()
```

When the raw data and the filtered data are plotted together we can see that the annual variations are increased over summer times, while winter times are less varied.  This makes sense as summer temperature go from cool nights to very warm days due to higher sun angle thus more direct sun and longer days thus more sun period and in the winter variations would not be so larg eas the sun angle is lower meaning less heating during the day and since less heat during the day less variation in temperature from night to day.
