## Q1: Distribution of power spectral estimates

<div style='background:#F0F0F0'>**1** You **may** use `matplotlib.mlab.psd` for the following question, but you had best use it correctly!

Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:  
</div>

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.
   2. Show that block averaging with no overlap is distributed as $\chi^2_{2N_{blocks}}$.
   3. Show that block averaging with 50% overlap Hanning windows is distributed as $\chi^2_{2N_{blocks}}$.

<div style='background:#F0F0F0'>You can choose your block length, but making it an integer divisor of 2048 will make your life a lot easier.  Show that as $N_{blocks}$ is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter).</div>


<div style='background:#F0F0F0'>HINT: for this time series the individual frequency estimates are indipendent samples of the distribution, so you can use them in compiling your distributions.</div>

<div style='background:#F0F0F0'>HINT: To get the pdf of the $\chi^2_\nu$ distribution correct, you need to "scale" by $\nu^{-1}$ where $\nu$ are the degrees of freedom.</div>

<div style='background:#F0F0F0'>HINT: for good presentation, make your histograms have the same bin sizes, and compare the distributions for all three cases on the same plot.</div>

```python
fig = plt.figure(figsize=(10, 5))
T = 2048
N = 2048
nblock = 8
dfrraw = 2.
dfrblock = 2.*nblock
dfrblockwind = 2.*((nblock*2)-1)
dfrblockwind1811 = (18./11.)*((nblock*2)-1)
dt = T/N
nfft = N/nblock
#standard case
def spectralestimate(x, T, nfft, dt, overlap):
    # spectral estimate
    Graw, f = mlab.psd(x, Fs = 1./dt, NFFT = nfft, noverlap=overlap)
    Graw = Graw.flatten()
    return Graw
#window case
def spectralestimate2(x, T, nfft, dt, overlap):
    # spectral estimate
    Graw, f = mlab.psd(x, Fs = 1./dt, NFFT = nfft, window = mlab.window_hanning, noverlap=overlap)
    Graw = Graw.flatten()
    return Graw
ax = fig.add_subplot(1, 1, 1)
Gvalr = np.zeros((100, N/2))
Gvalb = np.zeros((100, nfft/2))
Gvalbw = np.zeros((100, nfft/2))
Gvalbw1811 = np.zeros((100, nfft/2)) 
for i in range(0, 100, 1): 
    x = np.random.randn(N)
    Graw = spectralestimate(x, T, N, dt, 0)
    Gblock = spectralestimate(x, T, nfft, dt, 0)
    Gblockwind = spectralestimate2(x, T, nfft, dt, nfft/2.)
    Gblockwind1811 = spectralestimate2(x, T, nfft, dt, nfft/2.)
    Graw = Graw[1:]/2
    Gblock = Gblock[1:]/2
    Gblockwind = Gblockwind[1:]/2
    Gblockwind1811 = Gblockwind1811[1:]/2
    t = np.linspace(0, 10, 1000)
    interraw = stats.chi2.pdf(t, df = dfr, scale = 1/dfr)
    interblock = stats.chi2.pdf(t, df = dfrblock, scale = 1/dfrblock)
    interblockwind = stats.chi2.pdf(t, df = dfrblockwind, scale = 1/dfrblockwind)
    interblockwind1811 = stats.chi2.pdf(t, df = dfrblockwind1811, scale = 1/dfrblockwind1811)
    Gvalr[i,:] = Graw
    Gvalb[i,:] = Gblock
    Gvalbw[i,:] = Gblockwind
    Gvalbw1811[i,:] = Gblockwind1811
Gvalr = Gvalr.flatten()
Gvalb = Gvalb.flatten()
Gvalbw = Gvalbw.flatten()
Gvalbw1811 = Gvalbw1811.flatten()

#plots
ax.hist(Gvalr, bins = 40, label = r'$G_{raw}$', normed = True, histtype= 'step')
ax.plot(t, interraw, label = r'$\chi_{2}^{2}$')

ax.hist(Gvalb, bins = 40, label = r'$G_{block}$', normed = True, histtype= 'step')
ax.plot(t, interblock, label = r'$\chi_{2N_{blocks}}^{2}$')

ax.hist(Gvalbw, bins = 40, label = r'$G_{block window and overlap}$', normed = True, histtype= 'step')
ax.plot(t, interblockwind, label = r'$\chi_{2N_{blocks}}^{2}$ with window and overlap')

ax.set_xlabel ('Frequency [Hz]')
ax.set_ylabel(r'Spectral Estimate[$V^2$s]')
ax.set_title('Spectral Estimate Comparisons with $\chi^{2}$ Distributions')
ax.legend()


```

As $N_{blocks}$ increases the mean of the distribution of the raw set of $\chi_{2}^{2}$ and $G_{raw}$ is compared to the distribution of $\chi_{2N_{Blocks}}^{2}$ andd $G_{block}$.  It can be seen that the distribution narrows and the peak increases.  When we add an overlap thus increasing the Number of Blocks again the results again narrow and the mean(peak) occurs more frequently.  Narrowing curves mean that the variation from the mean is less and thus we can say that as $N_{blocks}$ increase the variance of the data decreases.

<div style='background:#F0F0F0'>**2**  For the Hanning window estimate, what fits better $2N_{blocks}$ degrees of freedom or $18N_{blocks}/11$ degrees of freedom (the correct theoretical value)?  Its OK to evaluate by eye.</div>

```python
# Note code for the theoretical situation is posted above in the code that runs each situation.  Please see above for the 18/11 code case results
fig = plt.figure(figsize=(10, 5))
ax= fig.add_subplot(1, 1, 1)
# 2N blocks
ax.hist(Gvalbw, bins = 40, label = r'$G_{block window and overlap}$', normed = True, histtype= 'step')
ax.plot(t, interblockwind, label = r'$\chi_{2N_{blocks}}^{2}$ with window and overlap')
# 18N blocks/11
ax.hist(Gvalbw1811, bins = 40, label = r'$G_{block window and overlap for 18/11 N}$', normed = True, histtype= 'step')
ax.plot(t, interblockwind1811, label = r'$\chi_{(18/11)N_{blocks}}^{2}$ with window and overlap')
ax.set_xlabel ('Frequency [Hz]')
ax.set_ylabel(r'Spectral Estimate[$V^2$s]')
ax.set_title('Spectral Estimate Comparisons with $\chi^{2}$ Distributions')
ax.legend()
```

The $\chi_{2N_{blocks}}^{2}$ fits the data better in both cases of the spectral estimates on our data set.  The 2N$_{blocks}$ fits better than the theoretical value of 18/11N$_{blocks}$.

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[np.isfinite(dc)]
```

```python
#hourly data
#the higher the nfft the more abrupt the spkies are.  Tried it at 200, then 2000.
nfft =2000
dcpsd, f = mlab.psd(dc, Fs = 1./(1.*60.*60.), NFFT =nfft, window=mlab.window_none, noverlap = (0.5*nfft))
dcpsdwin, fwin = mlab.psd(dc, Fs = 1./(1.*60.*60.), NFFT =nfft, window=mlab.window_hanning, noverlap = (0.5*nfft))
fig = plt.figure(figsize = (10, 5))
ax1 = fig.add_subplot(1, 1, 1)
ax1.loglog(dcpsd, label = 'Deep Cove Power Spectrum Normalized without Hanning Window and 50% overlap')
ax1.loglog(dcpsdwin, label = 'Deep Cove Power Spectrum Normalized with Hanning Window and 50% overlap')
ax1.set_xlabel ('Frequency [Hz]')
ax1.set_ylabel(r'Spectral Estimate[$^\circ$$C^2$s]')
ax1.set_title('Spectral Estimate of Deep Cove Data')
ax1.legend()
```

Spectral leakage is a result of the peaks combining together.  With windowing the peaks are more defined as the spectral leakage is reduced. When looking at our data set the blue line is without windowing and we can see at frequencies below $10^{2}$ that the green line (without hanning window) is smoother and peask are less pronouced in comparison to the hanning window (blue line) case. At frequencies of $10^{2}$ to $10^{3}$green peaks are more defined and blue peaks appear to be higher or wider.

<div style='background:#F0F0F0'> **2** Now do the same thing to the intergal of the Deep Cove data with time: $$y(t)=\int_o^t x(t')\ \mathrm{d}t'$$  This is obviously a silly thing to do, but compare the spectra and comment on the difference between the Hanning window and the non-Hanning windowed data.  </div>

```python
y = np.cumsum(dc)
ypsd, f = mlab.psd(y, Fs = 1./(1.*60.*60.), NFFT =nfft, window=mlab.window_none, noverlap = (0.5*nfft))
ypsdwin, fwin = mlab.psd(y, Fs = 1./(1.*60.*60.), NFFT =nfft, window=mlab.window_hanning, noverlap = (0.5*nfft))
fig = plt.figure(figsize = (10, 5))
ax1 = fig.add_subplot(1, 1, 1)
ax1.loglog(ypsd, label = 'Deep Cove Power Spectrum Normalized without Hanning Window and 50% overlap')
ax1.loglog(ypsdwin, label = 'Deep Cove Power Spectrum Normalized with Hanning Window and 50% overlap')
ax1.set_xlabel ('Frequency [Hz]')
ax1.set_ylabel(r'Spectral Estimate[$^\circ$$C^2$s]')
ax1.set_title('Spectral Estimate using the Integral of the Deep Cove Data over the time')
ax1.legend()
```

The spectra that has been hanning windowed can be clearly seen to have peaks while the resolution without the hanning window of the data can be seen to be small.  Small enough to the point that we can't see any peaks. In this case the Hanning window is key to observe actual data in the deep cove data set.
