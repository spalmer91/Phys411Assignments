## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.
<hr>

```python
def Corr(a):
    N = 200000
    x0 = np.random.randn(N)
    x_avg = np.mean(x0)
    x = x0 - x_avg
    x_stdev = np.std(x, ddof = 1)

    n = np.logspace(-2, 2., num = 40, endpoint = True, base = 10.0)
    n_inversenorm = (a*np.sum(x_stdev)/n)
    fig = plt.figure(figsize= (10, 5))
    ax = fig.add_subplot(1,1,1)
    r_values = np.zeros(40)
    y = np.arange(0., N, 1)
    for j in range (0, 40, 1):
        n_sub_i = np.random.normal(0,n[j],N)
        y = a*x+n_sub_i
        y_avg = np.mean(y)
        y_stdev = np.std(y, ddof = 1)
        y = y - y_avg
        r_values[j] =np.sum(x*y)/x_stdev/y_stdev/(N-1)
    ax.loglog(n_inversenorm, r_values)
    ax.set_xlabel ("Inverse Normalized Noise")
    ax.set_ylabel('Correlation Coefficient')
    ax.set_title('Correlation Coefficient with respect to Inverse Normalized Noise Amplitude:%s'%(a))
    
a1 = Corr(1.0)
```

As the noise is decreased the correlation between x and y increases to a maximum of a correlation value of 1.

```python
a2 = Corr(0.1)
```

As we in decrease the noise though the amplitude starts to dominate and our correlation becomes stronger approaching 1, the maximum value that can be attained.

## Q2: Monte Carlo on slopes and test versus student-t distribution

**1** We stated without proof that the slope $b$ of a linear fit of $y$ to $x$ is distributed as a student-t distribution with $N-2$ degrees of freedom, with a standard devaition given by the standard deviation of $y$ around the fit model $\tilde{y}$.   Lets test that emperically using a Monte Carlo technique.  

First create an independet variable $x$ from 10 points from an $N(0,1)$ distribution.

Next, using $n=1.0$, and $a=1.0$ create the dependent variables $y_i=a x_i + n_i$ where $n_i$ are drawn from the $N(0,n)$ distribution.  Calculate the slope $b$.  

Do this 10000 times, and make a histogram of the resulting fits $b$.  Also histogram $s_{y|x}$ as defined in the notes.

Comment on the histigrams

<hr>

```python
repeat = 10000
N = 10
a = 1.0
n = 1.0
bins=30
barray = np.zeros(repeat)
sy1x1array = np.zeros(repeat)
for i in range(0, repeat, 1):  
    x1 = np.random.normal(0, 1, N)
    n_i = np.random.normal(0, n, N)
    y1 = a*x1+n_i
    x1_mean = np.mean(x1)
    y1_mean = np.mean(y1)
    b = (np.sum(x1*y1) - (N*x1_mean*y1_mean))/(np.sum(x1**2)-(N*x1_mean**2))
    a1 = np.mean(y1) - (b*np.mean(x1))
    yhat = b*x1+a1
    sy1x1 = np.sqrt((np.sum((y1-yhat)**2))/(N-2))
    barray[i] = b
    sy1x1array[i] = sy1x1
fig = plt.figure(figsize = (10, 5))
ax1=fig.add_subplot(1, 1, 1)
ax1.hist(barray,bins, label = 'b')
ax1.set_title('Histogram of b values')
ax1.set_xlabel('B values')
ax1.set_ylabel('Occurences')
fig = plt.figure(figsize = (10, 5))
ax2=fig.add_subplot(1, 1, 1)
ax2 = fig.add_subplot(1, 1,1)
ax2.hist(sy1x1array,bins, label = 'Sample Standard Deviation, Sx|y')
ax2.set_title('Histogram of Sample Standard Deviation')
ax2.set_xlabel('Sample Standard Deviation')
ax2.set_ylabel('Occurences')
```

Both histograms have similar shape in that they both appear to be gaussian like.  As show in the next plots they are closer to the student t-distribution. The first plotappears to have a mean of 1 and a width (standard deviation) of 1 as well.  THe plot is fairly symmetric about the mean. The second plot displays the standard deviation values.  The standard deviation mean is approximately 1.  This shows that the approximation of a standard deviation inthe first graph of 1 is accurate.  This plot is also fairly symmetric about the mean with a slight skew towards the right hand side of the graph.  

**2** Assuming $s_{y|x}=1.$ (which is the same as our value of $n$), plot the student-t probability density function on the histogram of b and show that they compare favorably (if they don't you might need to check your work!)

<hr>

```python
fig = plt.figure(figsize = (10, 5))
ax1=fig.add_subplot(1, 1, 1)
ax1.hist((barray-np.mean(barray))/np.std(barray),bins, label = 'b', normed = 1)
ax1.set_title('PDF of b values')
ax1.set_xlabel('B values Centralized (with mean removed)')
ax1.set_ylabel('Occurences of B Normalized')
t = np.linspace(-6,6, 10000)
s = stats.t.pdf(t,10000, scale = 1.0)
ax1.plot(t, s,label = 'Student t-Distribution' )
ax1.legend()
```

## Q3: Linear comparison of Deep Cove data to James Bay temperature data

```python
# Load the data:

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb=hourdata[1,:]
```

**1** Make a scatterplot of Plot James Bay versus Deep Cove.

```python
fig = plt.figure(figsize = (10,5))
ax4 = fig.add_subplot(1, 1, 1)
ax4.scatter(jb, dc)
ax4.set_title('James Bay vs Deep Cove')
ax4.set_ylabel('Deep Cove Data')
ax4.set_xlabel('James Bay Data')
```

**2** Compute the linear Correlation Co-efficient.

```python
good = np.isfinite(dc+jb)
dc_clean = dc[good]
jb_clean = jb[good]
R2=(np.sum((dc_clean-np.mean(dc_clean))*(jb_clean-np.mean(jb_clean)))**2)/(np.sum((dc_clean-np.mean(dc_clean))**2))/(np.sum((jb_clean-np.mean(jb_clean))**2))
print ('Linear Correlation = '), R2
```

**3** Fit a straight line to the James Bay data as a fucntion of the Deep Cove data.  Comment on the fit.  (Do **NOT** use a canned routine for the fitting)

<hr>

```python
dc_cleanadj=dc_clean - np.mean(dc_clean)
jb_cleanadj=jb_clean - np.mean(jb_clean)
b = (np.sum(dc_cleanadj*jb_cleanadj))-(len(dc_cleanadj)*np.mean(dc_cleanadj)*np.mean(jb_cleanadj))
b2 = b/(np.sum((dc_cleanadj)**2)-N*np.mean(dc_cleanadj)**2)
a = np.mean(jb_cleanadj)-b*np.mean(dc_cleanadj)
yhat=b2*dc_cleanadj+a
fig = plt.figure(figsize = (10, 5))
ax= fig.add_subplot(1,1,1)
ax.scatter(dc_cleanadj,jb_cleanadj)
ax.plot(dc_cleanadj,yhat,'-',label='Fit: y=%1.4f+%1.4fx'%(a,b2))
ax.set_title('James Bay vs Deep Cove')
ax.set_ylabel('James Bay Data')
ax.set_xlabel('Deep Cove Data')
ax.legend()
```

For a proper fit the data should be aproximately evenly distributed above and below the fit line.  In this fit it appears the the line seems to split the data in center.

**4** Add the confidence intervals to your fit, and comment.  ***NO CANNED ROUTINES***

```python
dc_cleanadj=dc_clean - np.mean(dc_clean)
jb_cleanadj=jb_clean - np.mean(jb_clean)
b = (np.sum(dc_cleanadj*jb_cleanadj))-(len(dc_cleanadj)*np.mean(dc_cleanadj)*np.mean(jb_cleanadj))
b2 = b/(np.sum((dc_cleanadj)**2)-N*np.mean(dc_cleanadj)**2)
a = np.mean(jb_cleanadj)-b*np.mean(dc_cleanadj)
yhat=b2*dc_cleanadj+a
fig = plt.figure(figsize = (10, 5))
ax5=fig.add_subplot(1,1,1)
ax5.scatter(dc_cleanadj,jb_cleanadj)
ax5.plot(dc_cleanadj,yhat,'-',label='Fit: y=%1.4f+%1.4fx'%(a,b2))
ax5.set_title('James Bay vs Deep Cove')
ax5.set_ylabel('James Bay Data')
ax5.set_xlabel('Deep Cove Data')
ax5.legend()

sxy = np.sqrt(np.sum((jb_cleanadj-yhat)**2)/(N-2))
bdn, bup = stats.t.interval(0.95, df=N, scale = sxy)
bb = np.array([bdn, bup])
bb = bb/np.sqrt(np.sum((dc_cleanadj-np.mean(dc_cleanadj))**2))
xx=np.sort(dc_cleanadj)
yhathigh = (b2+bb[1])*(xx-np.mean(dc_cleanadj))+np.mean(jb_cleanadj)
yhatlow = (b2+bb[0])*(xx-np.mean(dc_cleanadj))+np.mean(jb_cleanadj)
ax5.fill_between(xx,yhathigh,yhatlow,alpha=0.5)
```

The confidence interval is within the range of the data still and thus the variation in the slope to the upper or lower bound would create aslope that would adequately describe the data plotted.
