**1** Using the data from the hourly time series at Deep Cove (`dc`), show that the canned `fft` routine returns the correct discrete Fourier Transform of a suitable subset of the data (your choice on what subset).  You can do this graphically, by comparing the amplitude and the phase of the Fourier components, or simply show that the resulting numbers are the same.  

Hints: 
  - ignore bad data (rather than dealing with it "properly").  Make sure the units of your FT are: $\mathrm{^oC\, s}$.
  - If you want to compute the DFT effeciently, form the matrix $F$ referred to in the notes, and plot some of the time series associated with the different entries and make sure they look like Sines and Cosines and that the edge conditions are what you want


````python
good = np.isfinite(dc)
dcclean = dc[good]
dcsubset = dcclean[0:500]
#dt: sampling rate (normally 1)
def dft(x, dt):
    T = len(x)
    N = len(x)/dt
    cos = np.zeros(N)
    sin = np.zeros(N)
    k = N/2
    real = np.zeros(k)
    imag = np.zeros(k, dtype = complex)
    X = np.zeros(k, dtype = complex)
    fk = np.zeros(k)
    #k goes to N/2
    for kk in range(0, k, 1):
        for n in range(0, N, 1): 
            cos[n] = np.cos((2*np.pi*kk*n)/N)
            sin[n] = np.sin((2*np.pi*kk*n)/N)
        real[kk] = np.sum(x*cos)
        imag[kk] = np.sum(x*sin)
        X[kk] = real[kk]+(-1j*imag[kk])
        fk[kk] = kk
    fk=fk/T
    X = T*X/N
    return X, fk
X, fk = dft(dcsubset, 1)
fig = plt.figure(figsize = (10, 5))
ax3=fig.add_subplot(1,1,1)
ax3.loglog(fk, np.abs(X))
ax3.set_xlabel ('Frequency [Hz]')
ax3.set_ylabel(r'Fourier Transform [$^\circ$$C^2$s]')
ax3.set_title('Definition Deep Cove Fourier Transfrom')
````

````python
DCX = np.fft.fft(dcsubset)
fdcx = np.fft.fftfreq(len(dcsubset))
fig = plt.figure(figsize = (10, 10))
ax=fig.add_subplot(2,1,1)
ax.loglog(fdcx, np.abs(DCX))
ax.set_xlabel ('Frequency [Hz]')
ax.set_ylabel(r'Fourier Transform [$^\circ$$C^2$s]')
ax.set_title('Numpy Deep Cove Fourier Transfrom')
ax1 = fig.add_subplot(2,1,2)
ax1.loglog(fdcx, np.abs(DCX), label = "Numpy FFT function")
ax1.loglog(fk, np.abs(X), label = "Definition FFT funtion")
ax1.set_xlabel ('Frequency [Hz]')
ax1.set_ylabel(r'Fourier Transform [$^\circ$$C^2$s]')
ax1.set_title('Deep Cove Fourier Transfrom Comparison Between Numpy and Definition FFT Function')
ax1.legend()
````

The first plot was created using the definition of the fourier transform with sines and cosines as defined in the notes.  The second plot was created using the numpy fuction for fast fourier transforms.
When the two outputted fourier transforms were plotted and overlaid on one another the two plots are the same thus the definition of the fourier transform is the numpy's version of the fourier transform.  Thus we can use the fourier transforms np.fft.fft.

**1** Write a short function to create a "red" time series by taking the cumulative sum of Normally distributed random data, and plot one realization.  The number of data points $N$ can be relatively small (say 1000).

````python
def culm(N):
    x = np.random.randn(N)
    y = np.cumsum(x)
    n = np.arange(0, len(y), 1)
    return y, n
x, n = culm(1000)
fig = plt.figure(figsize = (10, 5))
ax=fig.add_subplot(1,1,1)
ax.plot(n, x)
ax.set_xlabel ('Number of Data Points')
ax.set_ylabel('Cumulative Sum')
ax.set_title('Cumulative Sum of Normally Distirbuted Random Data')
````

**2** Write a second function to return the estimate of the power spectrum of $x$.  For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate.  Plot the resulting spectrum.

````python
def ps():
    x = np.cumsum(np.random.randn(1000))
    z = np.fft.fft(x)
    f = np.fft.fftfreq(len(x))
    sxx = (np.conj(z)*(z))/len(z)
    return sxx, f
sxx, f = ps()  
fig = plt.figure(figsize = (10, 5))
ax=fig.add_subplot(1,1,1)
ax.loglog(f, np.real(sxx))
ax.set_xlabel ('Frequency [Hz]')
ax.set_ylabel(r'$S_{xx}$ [$V^2$s]')
ax.set_title('Deep Cove Power Spectrum')
````
