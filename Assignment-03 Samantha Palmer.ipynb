## Q1: Lag correlations of weather time series

````python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb=hourdata[1,:]
````

a**1** From the hourly time series at Deep Cove (`dc`), calculate the *lag correlation co-efficient* $\rho_{xx}(\tau)$ out to a lag of $\tau=30\ \mathrm{days}$ (Umm, remember this data is hourly!).  (also rememebr to remove the mean of $dc$ before trying to to the calculations).  

Comment on the resulting plot.

````python
good = np.isfinite(dc)
dcclean = dc[good]
dcdata = dcclean - np.mean(dcclean)
lags = np.arange(0, 720.)
cxx = 0.*lags
stddc = np.std(dcdata)
rhoxx = 0.*lags
for i, tau in enumerate(lags):
    if tau==0:
        cxx[i]=np.mean(dcdata*dcdata)
    else:
        cxx[i]=np.mean(dcdata[:-tau]*dcdata[tau:])
    rhoxx[i] = cxx[i]/(stddc*stddc)
fig = plt.figure(figsize = (10, 5))
ax1=fig.add_subplot(1,1,1)
ax1.plot(lags, rhoxx )
ax1.set_xlabel (r'$\tau (h))$')
ax1.set_ylabel(r'$\rho_{xx}$')
ax1.set_title('Deep Cove Lag Coorelation Coefficient')
````

The lag coorelation has a decreasing trend as tau becomes larger. The more lag introducedup to 30 days the less the lag correlation. The frequency observed has a even distribution and thus could be a result of daily coorelations.

*2* Now, instead of doing the whole year, look at just the winter months and the summer months and compare the lag correlations (no need to do this for the whole time series, just choose one winter and one summer).  Make sure to indicate which lag correlation is which on the plots.

Comment below on the different charcter of these two lag correlations.

````python
#start date January 1st midnight
#winter: Dec 21st - March 19
#summer: June 21 - Sept 22
dcwinter = dc[355*24:(355+88)*24]
dcsummer = dc[172*24:(172+93)*24]
goodw = np.isfinite(dcwinter)
dcwinterclean = dcwinter[goodw]
goods = np.isfinite(dcsummer)
dcsummerclean = dcsummer[goods]
dcwinterdata = dcwinterclean - np.mean(dcwinterclean)
dcsummerdata = dcsummerclean - np.mean(dcsummerclean)

lags = np.arange(0, 720.)
cxx1 = 0.*lags
cxx2 = 0.*lags
stddwc = np.std(dcwinterdata)
stddsc = np.std(dcsummerdata)
rhoxxwinter = 0.*lags
rhoxxsummer = 0.*lags
for i, tau in enumerate(lags):
    if tau==0:
        cxx1[i]=np.mean(dcwinterdata*dcwinterdata)
        cxx2[i]=np.mean(dcsummerdata*dcsummerdata)
    else:
        cxx1[i]=np.mean(dcwinterdata[:-tau]*dcwinterdata[tau:])
        cxx2[i]=np.mean(dcsummerdata[:-tau]*dcsummerdata[tau:])
    rhoxxwinter[i] = cxx1[i]/(stddwc*stddwc)
    rhoxxsummer[i] = cxx2[i]/(stddsc*stddsc)
fig = plt.figure(figsize = (10, 5))
ax1=fig.add_subplot(1,1,1)
ax1.plot(lags, rhoxxwinter, label = 'Winter Lag' )
ax1.plot(lags, rhoxxsummer, label = 'Summer Lag')
ax1.set_xlabel (r'$\tau (h)$')
ax1.set_ylabel(r'$\rho_{xx}$')
ax1.set_title('Deep Cove Lag Coorelation Coefficient')
ax1.legend()
````

When looking at the coorelation initially both systems decrease in lag coorelation. The summer coorelation then levels out after an amount of lag.  The frequency for the summer remains the same and thus could show daily fluctuation.  The winter month takes much longer to level out and does not become generally level.  Its mean amplitude of coorelation changes throughout the lag.  This could mean there are more vairations of temperature from the data in winter months then in summer months.

**1** Consider the process $x(t) = \int_{t-b}^t y(t) \ \mathrm{dt}$ where $y(t)$ is uncorrelated white noise so that $R_{yy}(\tau)=\delta(\tau)$.  Compute $R_{xx}(\tau)$.

**2** Show emperically that you have the right *lag auto correlation* by computing the lag correlation of a random discrete timeseries $x(t)$.  Hint: have a look at `np.convolve` to do the integration. Hint 2: Make sure your time series is long enough to get a nice representation of $R_{xx}(\tau)$.  Show what happens if N is too small and what happens if it gets larger.
(Hint to do this well, I needed N to get pretty large, like 5e6,  I used `N in [500, 5000, 50000, 5e6]`, but my machine has a good amount of memory.  Don't *test* on the bigger values, just add them for the final product).

````python
fig = plt.figure(figsize = (10, 5))
def lagcor(N,T,b):
    y = np.random.randn(N+b)
    x = np.zeros(N)
    for i in range(0,N): 
        x[i] = np.sum(y[i:i+b])  
    lags = np.arange(0, T)
    rxx = 0.*lags
    for i, tau in enumerate(lags):
        if tau==0:
            rxx[i]=np.mean(x*x)/np.var(x,ddof=1)
        else:
            rxx[i]=np.mean(x[:-tau]*x[tau:])/np.var(x,ddof=1)
    ax1=fig.add_subplot(1,1,1)
    ax1.plot(lags, rxx, label = 'N = %s'%(N))
    ax1.set_xlabel (r'$\tau (s)$')
    ax1.set_ylabel(r'$\ R_{xx}$')
    ax1.set_title('Lag Coorelation')
    ax1.legend()
    return rxx
T = 100
b = 30
rxx1 = lagcor(500,T,b)
rxx2 = lagcor(5000, T,b)
rxx3 = lagcor(50000, T,b)
rxx4 = lagcor(5000000, T,b)
````

````python
fig = plt.figure(figsize = (10, 5))
dt = 1.0
f = np.arange(0., 0.5, 0.001)
def dft(x,dt, f, N): 
    # where x is your time series, dt is the sample rate in seconds, 
    # and f are the frequencies in Hz you want the Fourier Series calculated on
    n = len(x)
    X = np.zeros(len(f))
    XF = np.zeros(len(f))
    Tarray = np.arange(0., n, 1.)
    for fi in range (0, len(f)):
        X[fi] = np.sum(x*np.exp(-1j*2*np.pi*f[fi]*Tarray)) #contains imaginary and real components
        XF[fi] = (np.sqrt((np.real(X[fi])**2)+(np.imag(X[fi])**2)))
    ax1=fig.add_subplot(1,1,1)
    ax1.plot(f, XF, label = 'N = %s'%(N))
    ax1.set_xlabel ('f (Hz)')
    ax1.set_ylabel(r'Log($\ S_{xx}$) ($\ V^2$/Hz)')
    ax1.set_yscale('log')
    ax1.set_title('Power Spectral Density')
    ax1.legend()
    return XF
XF1 = dft(rxx1, dt, f, 500)
XF2 = dft(rxx2, dt, f, 5000)
XF3 = dft(rxx3, dt, f, 50000)
XF4 = dft(rxx4, dt, f, 5000000)
````
