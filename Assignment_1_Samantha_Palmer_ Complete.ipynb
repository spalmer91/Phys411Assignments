## Q1: Simulate aliasing of sinusoidal signals

$$ x = A_1 \cos (\omega t) + A_2 \cos ((\omega+\delta \omega ) t) $$

where $\omega$ is the "carrier" frequency, and $\delta \omega << \omega$ is the difference in frequencies between the two waves.  For simplicty, you could make $A_1=A_2=1.0\ [V]$.  Feel free to choose an interesting $\omega$.

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{flip}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll.  
<hr>

```python
def flipit(n,m,seedit):
    pH = 2./3.
    pT = 1./3.
    h= 0
    t = 0
    x = np.random.seed(seedit)
    x = np.random.rand(m, n)
    for j in range (0, m):
        for i in range (0, n):
            if x[j, i]<pH:  
                x[j,i]=0.5
            else:
                x[j,i]=-0.5
        mean = np.mean(x, axis=1)  
        variance= np.var(x, axis=1)
    return mean
NOTest = 2000
seed = 865
x2 = flipit(2, NOTest, seed)
x16 = flipit(16, NOTest, seed)
x64 = flipit(64, NOTest, seed)
x256 = flipit(256, NOTest, seed)
bins = 30
fig = plt.figure(figsize= (10, 10))
ax3 = fig.add_subplot(2,1,1)
ax3.hist(x2,bins, label ='Mean for 2 Flips')
ax3.hist(x16,bins, label = 'Mean for 16 Flips')
ax3.hist(x64,bins, label = 'Mean for 64 Flips')
ax3.hist(x256,bins, label = 'Mean for 256 Flips')
ax3.set_xlabel('Mean')
ax3.set_ylabel('Occurences')
ax3.set_title('Sample Means')
ax3.legend(loc=1)
```

As the number of flips is increased it can be seen above the centrality of the flips becomes more focused at the expected mean of the means of flips of 0.167.  The initial flip of 2 is very broad and as we increase the number of flips the means become closer together and concentrated as can be seen in the mean for 256 flips.  When comparing these results to the convergence of the sample mean of the dice roll we can see that initially the lower number of rolls is spread across the graph. The data is similar in height and does appear to have a unique peak to define the means mean.  When looking at the low number of flips the data is also very well spread with no distinct peak.  As the number of rolls increases the data becomes more focused at the mean of the means.  Each time the mean of means peak becomes more central around the convergent point.  There becomes a single peak zone that can be seen in each increase of rolls. 
This peak data zone can also be seen in the flip data above.  As we increase the number of flips the data becomes more concentrated around the mean of the means and the peak becomes more established when compared to the surrounding data.  Thus in both data sets we can see that as the number of rolls increase the apparent mean of the means has a more concentrated set of data around the mean of the means. With each increase in fliprs or rolls the data converges to the mean of means and thus to the expected mean.

**5** Replot the histograms, removing the mean and normalizing by the standard deviation.  Also show the CDF for normal distribution with mean zero and standard deviation of 1.
<hr>

```python
bins=40
allmeans=[x2, x16, x64, x256]
labl = ['2 Flips','16 Flips','64 Flips','256 Flips']
fig = plt.figure(figsize= (20, 6))
ax4 = fig.add_subplot(1,1,1)
histos=list()
for i, j in enumerate(allmeans):
    histo, bins, patches = ax4.hist(((j-np.mean(j))/np.std(j)), bins, label = labl[i], histtype='step', normed=True)
    histos.append(histo)
ax4.set_xlabel('Mean Flip')
ax4.set_ylabel('PDF [per flip increment]')
ax4.set_title('Sample Means')
midbins = bins[0:-1] - np.diff(bins)[0]
cdf = stats.norm.cdf(midbins, 0, 1.)
ax4.plot(midbins,cdf, 'k--', label='CDF Normal Distribution')
ax4.legend(loc=1)
ax4.set_xlim(-2, 2)
```
